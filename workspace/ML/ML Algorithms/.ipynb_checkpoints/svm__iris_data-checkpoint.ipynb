```python
import import_ipynb
from core_data import *

import numpy as np
import pandas as pd

from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

---

#### Обучаем модель

1. Стандартизированные данные

```python
svm_std = SVC(kernel='linear', C=1.0, random_state=1)
svm_std.fit(X_train_std, y_train)
```

2. Нестандартизированные данные

```python
svm = SVC(kernel='linear', C=1.0, random_state=1)
svm.fit(X_train, y_train)
```

---

#### Прогноз на тестовой выборке

1. Стандартизированные данные

```python
y_svm_str = svm_std.predict(X_test_std)
print('Точность: %.3f' % accuracy_score(y_test, y_svm_str))
```

2. Нестандартизированные данные

```python
y_svm = svm.predict(X_test)
print('Точность: %.3f' % accuracy_score(y_test, y_svm))
```

---

#### Визуализация области решений

1. Стандартизированные данные

```python
plot_decision_regions(
    X_combined_std,
    y_combined,
    classifier=svm_std,
    test_idx=range(105, 150)
)
plt.xlabel('Длина лепестка [нестандартизованная]')
plt.ylabel('Ширина лепестка [нестандартизованная]')
plt.tight_layout()
plt.show()
```

2. Нестандартизированные данные

```python
plot_decision_regions(
    X_combined,
    y_combined,
    classifier=svm,
    test_idx=range(105, 150)
)
plt.xlabel('Длина лепестка [нестандартизованная]')
plt.ylabel('Ширина лепестка [нестандартизованная]')
plt.tight_layout()
plt.show()
```
