```python
import import_ipynb
from core_data import *

import matplotlib.pyplot as plt

import numpy as np

from sklearn.tree import DecisionTreeClassifier, plot_tree
```

---

#### Обучаем модель

1. Стандартизированные данные 

```python
tree_model_std = DecisionTreeClassifier(criterion='gini',
                                        max_depth=4,
                                        random_state=1)
tree_model_std.fit(X_train_std, y_train)
```

2. Нестандартизированные данные

```python
tree_model = DecisionTreeClassifier(criterion='gini',
                                    max_depth=4,
                                    random_state=1)
tree_model.fit(X_train, y_train)
```

---

#### Область решений

1. Стандартизированные данные 

```python
plot_decision_regions(X_combined,
                      y_combined,
                      classifier=tree_model_std,
                      test_idx=range(105, 150))

plt.xlabel('Длина лепестка [см]')
plt.ylabel('Ширина лепестка [см]')
plt.tight_layout()
plt.show()
```

2. Нестандартизированные данные

```python
plot_decision_regions(X_combined,
                      y_combined,
                      classifier=tree_model,
                      test_idx=range(105, 150))

plt.xlabel('Длина лепестка [см]')
plt.ylabel('Ширина лепестка [см]')
plt.tight_layout()
plt.show()
```

```python
feature_names = [
    'Sepal len', 'Sepal wid',
    'Petal len', 'Petal wid'
]
plot_tree(tree_model, feature_names=feature_names, filled=True)
```
