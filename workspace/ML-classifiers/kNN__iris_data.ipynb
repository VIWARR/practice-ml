```python
import import_ipynb
from core_data import *

import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

---

#### Обучаем модель

1. Стандартизированные данные

```python
knn_std = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
knn_std.fit(X_train_std, y_train)
```

2. Нестандартизированные данные

```python
knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
knn.fit(X_train, y_train)
```

---

#### Визуализация области решений

1. Стандартизированные данные

```python
plot_decision_regions(
    X_combined_std,
    y_combined,
    classifier=knn_std,
    test_idx=range(105, 150)
)
plt.xlabel('Длина лепестка [нестандартизованная]')
plt.ylabel('Ширина лепестка [нестандартизованная]')
plt.tight_layout()
plt.show()
```

2. Нестандартизированные данные

```python
plot_decision_regions(
    X_combined_std,
    y_combined,
    classifier=knn,
    test_idx=range(105, 150)
)
plt.xlabel('Длина лепестка [нестандартизованная]')
plt.ylabel('Ширина лепестка [нестандартизованная]')
plt.tight_layout()
plt.show()
```

---

#### Прогноз на тестовой выборке

```python
y_pred_std = knn_std.predict(X_test_std)
print('Точность: %.3f' % accuracy_score(y_test, y_pred_std))
```
