```python
import import_ipynb
from core_data import *

import matplotlib.pyplot as plt

import numpy as np

from sklearn.svm import SVC
```

---

#### Генерируем массив XOR

```python
np.random.seed(1)
X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                       X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, 0)
```

```python
plt.scatter(
    X_xor[y_xor == 1, 0],
    X_xor[y_xor == 1, 1],
    c='blue', marker='o',
    label='Class 1'
)

plt.scatter(
    X_xor[y_xor == 0, 0],
    X_xor[y_xor == 0, 1],
    c='red', marker='s',
    label='Class 0'
)

plt.xlim([-3, 3])
plt.ylim([-3, 3])

plt.xlabel('Признак 1')
plt.ylabel('Признак 2')

plt.legend(loc='best')
plt.tight_layout()
plt.show
```

---

#### Обучаем модель

1. Данные XOR

```python
svm = SVC(kernel='rbf', random_state=1, gamma=0.2, C=10.0)
svm.fit(X_xor, y_xor)
```

2. Стандартизированные данные

```python
svm_std = SVC(kernel='rbf', random_state=1, gamma=0.2, C=10.0)
svm_std.fit(X_train_std, y_train)
```

3. Нестандартизированные данные

```python
svm_nonstd = SVC(kernel='rbf', random_state=1, gamma=0.2, C=10.0)
svm_nonstd.fit(X_train, y_train)
```

---

#### Визуализация области решений

1. Данные XOR

```python
plot_decision_regions(X_xor, y_xor, classifier=svm)
plt.show()
```

2. Стандартизированные данные

```python
plot_decision_regions(X_train_std, y_train, classifier=svm_std)
plt.show()
```

3. Нестандартизированные данные

```python
plot_decision_regions(X_train, y_train, classifier=svm_nonstd)
plt.show()
```
