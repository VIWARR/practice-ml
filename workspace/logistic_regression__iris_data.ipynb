```python
import import_ipynb

import preceptron_traning__iris_data

from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

import matplotlib.pyplot as plt
```

#### Обучаем модели логистической регрессии

1. Стандартизированные данные

```python
lr_std = LogisticRegression(C=100.0, solver='lbfgs')
ovr_classifier_std = OneVsRestClassifier(lr_std)

ovr_classifier_std.fit(
    preceptron_traning__iris_data.X_train_std,
    preceptron_traning__iris_data.y_train
)
```

2. Нестандартизированные данные

```python
lr = LogisticRegression(C=100.0, solver='lbfgs')
ovr_classifier = OneVsRestClassifier(lr)

ovr_classifier.fit(
    preceptron_traning__iris_data.X_train,
    preceptron_traning__iris_data.y_train
)
```

#### Прогноз на тестовой выборке

1. Стандартизированные данные

```python
y_lr_std = ovr_classifier_std.predict(preceptron_traning__iris_data.X_test_std)
print('Точность: %.3f' % accuracy_score(
    preceptron_traning__iris_data.y_test, y_lr_std
))
```

2. Нестандартизированные данные

```python
y_lr = ovr_classifier.predict(
    preceptron_traning__iris_data.X_test
)
print('Точность: %.3f' % accuracy_score(
    preceptron_traning__iris_data.y_test, y_lr)
     )
```

#### Визуализация области решений

1. Стандартизированные данные

```python
preceptron_traning__iris_data.plot_decision_regions(
    preceptron_traning__iris_data.X_combined_std,
    preceptron_traning__iris_data.y_combined,
    classifier=ovr_classifier_std,
    test_idx=range(105, 150)
)
plt.xlabel('Длина лепестка [нестандартизованная]')
plt.ylabel('Ширина лепестка [нестандартизованная]')
plt.tight_layout()
plt.show()
```

2. Нестандартизированные данные

```python
preceptron_traning__iris_data.plot_decision_regions(
    preceptron_traning__iris_data.X_combined,
    preceptron_traning__iris_data.y_combined,
    classifier=ovr_classifier,
    test_idx=range(105, 150)
)
plt.xlabel('Длина лепестка [нестандартизованная]')
plt.ylabel('Ширина лепестка [нестандартизованная]')
plt.tight_layout()
plt.show()
```
